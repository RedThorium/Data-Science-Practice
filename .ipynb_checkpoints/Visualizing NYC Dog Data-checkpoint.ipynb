```python
# Let's read our data and analyzing it
dogData = pd.read_csv("NYC_Dog_Licensing_Dataset.csv")
```

```python
# we have to get rid of all nan values so we use dropna() function
# but first replacing empty values with nan values so they can be dropped

dogData.replace(' ', np.nan, inplace=True)
dogData.dropna(inplace = True)
```

```python
dogData.info()
```

# Dog Gender Visulization

```python
# Global sales ratio
gender_list = list(dogData.AnimalGender.unique())
gender_count_list = []

for i in gender_list:
    x = dogData[dogData.AnimalGender == i]
    gender_count = int(x.AnimalGender.value_counts())
    gender_count_list.append(gender_count)
    print(gender_count)
```

```python

#data = pd.DataFrame({"platform_list": platform_list,"global_sales_ratio": global_sales_ratio})

print(gender_list)

# visualization
plt.figure(figsize=(10,10))
sns.color_palette("dark")
sns.barplot(x=gender_list, y=gender_count_list)
plt.title("Dog Gender Breakdown in NYC",fontsize=25)
plt.xlabel("Genders",fontsize=15)
plt.ylabel("Counts",fontsize=15)
plt.show()
```

```python
labels = dogData.AnimalGender.value_counts().index
colors = ["red","blue"]
explode = [0,0]
sizes = dogData.AnimalGender.value_counts().values
plt.figure(figsize=(10,10))
plt.pie(sizes, explode=explode,labels=labels, colors=colors, autopct='%1.1f%%')
plt.show()

print(dogData.AnimalGender.value_counts())
print(labels)
```

# Breed Visualization

```python
# Breed count
breed_count = Counter(dogData.BreedName)
data2 = breed_count.most_common(10)
breed,count= zip(*data2)
breed,count = list(breed),list(count)

#visualization
plt.figure(figsize=(30,30))
sns.barplot(x = breed, y = count, palette = sns.color_palette("Blues"))
plt.title("Frequency of Breeds with Unknown Breed",fontsize=25)
plt.xlabel("Breeds",fontsize=15)
plt.ylabel("Frequency",fontsize = 15)
plt.show()
```

```python
# Breed count
breed_count = Counter(dogData.BreedName)
data3 = breed_count.most_common(11)
breed2,count2= zip(*data3[1:])
breed2,count2 = list(breed2),list(count2)

#visualization
plt.figure(figsize=(30,30))
sns.barplot(x = breed2, y = count2, palette = sns.color_palette("Reds"))
plt.title("Frequency of Breeds without Unknown Breed",fontsize=25)
plt.xlabel("Breeds",fontsize=15)
plt.ylabel("Frequency",fontsize = 15)
plt.show()
```

# Borough Visualization

```python
# Borough count
borough_count = Counter(dogData.Borough)
data4 = borough_count.most_common(5)
borough, boroughCount = zip(*data4)
borough, boroughCount = list(borough), list(boroughCount)

plt.figure(figsize=(10,10))
sns.barplot(x = borough, y = boroughCount)
plt.title("Dogs By Borough",fontsize=25)
plt.xlabel("Boroughs",fontsize=15)
plt.ylabel("Frequency",fontsize = 15)
plt.show()
```
